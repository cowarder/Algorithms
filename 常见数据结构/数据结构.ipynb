各种排序算法的比较

1.插入排序：对第i个元素进行排序的时候默认前1-（i-1）个元素已经是排序好的，将第i个元素插入到前i-1个元素中，最好情况是已经排序完成的

2.希尔排序：将待排序的数据按照增量划分为几个增量分组，每个增量分组应用插入排序，逐步减少增量到1，但是这种方法不稳定，受到增量选择的影响
  上界是n^2，下界是n*log2n，属于第一批打破n^2的算法
 
3.冒泡排序：从第一个到倒数第二个元素，前面的元素逐个与后面的比较，大于后面的元素，就进行交换，最后遍历一次之后最后一个元素最大，然后倒数第二个   元素作为最后一个元素继续遍历

4.快速排序：通过一趟排序将数组分割为独立的两部分，一部分比记录的关键字的值都小，一部分比记录的关键字的值都大，重复对这两个部分进行上述操作，关   键在于partion(low,high)操作，返回一个pivot的坐标，此时对数组中low到high的元素已将根据pivot进行了分割
   优化：1.对枢轴的选择，选择的枢轴应该是大体位于中数位置的，使得两边切分的比较接近（三数取中的方法）
       2.优化不必要的交换，将pivot先保存起来，每次做的不是交换而是赋值
       3.减少递归
       4.对于小数组直接使用插入排序（小数组优化）

5.选择排序：对于第i个元素，每一次都选择i到n的元素中的最小值与第i个元素交换

6.堆排序：构建一个大顶堆，每次去掉最大元素，构建一个新的大顶堆，堆排序对于原始记录的并不敏感，所以初始排列无论什么情况下都是O(nlogn)
    主要的时间耗费在删除堆顶元素之后，维护堆的交换操作上

7.归并排序：假设起始序列含有n个记录，则可以看成是n个有序的子序列，两两进行归并，重复此操作，直到得到一个有序的序列
    merge中进行的是两两比较，不存在跳跃，所以它是稳定的

堆排序和快速排序的比较：
    1.平均时间上，堆排序的时间常数比快排要大一些，会慢一些
    2.堆排序poor use of cache，快排在进行递归方法时，会倾向于选择局部数据，因此内存的命中率会高一些，也就是说局部性上面快速排序更优
    3.如果观察堆排序的话，数据在数组中的位置是不断大幅度改变的，但是在快速排序中，数据只在小范围内进行摆动

树结构
满二叉树：叶子结点一定是在最下一层，所有的非叶子结点的度都为2
完全二叉树：叶子结点只能存在于最下两层，最下层叶子结点只能由左向右排列
满二叉树一定是完全二叉树，但是完全二叉树不一定是满二叉树

二叉树的性质：
具有n个结点的完全二叉树的深度为floor(log2底N)+1
